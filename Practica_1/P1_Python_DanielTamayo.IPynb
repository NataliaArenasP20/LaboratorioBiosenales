Procederemos a crear vectores y realizar operaciones en estos utilizando la librería numpy importada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Creación de vectores
a =np.array([3.1 , 1, -0.5, -3.2, 6]) 
b= np.array ([1, 3 , 2.2 , 5.1 , 1])

# Producto escalar con numpy
pe = a @ b

# Producto escalar punto a punto
pep = a[0]*b[0] + a[1]*b[1] + a[2]*b[2] + a[3]*b[3] + a[4]*b[4]

print ("Producto escalar np : "+str(pe), " producto escalar punto a punto : ",str(pep),"\n")

# Creación matriz A
A = np.array([[2, -1, -3], [4, 1.5, -2.5], [7.3, -0.9, 0.2]])
print("Matrices A y A transpuesta: \n",A," \n")
# A transpuesta
AT= np.transpose(A)
print(AT)
```

Ahora emplearemos numpy y matplotlib para crear y graficar funciones.

```python
# Creación intervalo de la función (Dominio) (0 <=n <= 100)
Dom = np.arange(0,101)

# Emplear la función
Y1 = np.sin(np.pi * 0.12 * Dom)

# Graficar la función
plt.plot(Dom, Y1,label='Y(n)')
plt.xlabel('n')
plt.ylabel('Y(n)')
plt.title('Gráfica de Y = sin(pi*0.12n)')
plt.grid(True)
plt.legend()
plt.show()

# Segunda Función
Y2 = np.cos(2*np.pi * 0.03 * Dom)
plt.plot(Dom, Y2, label='Y(n)',color='orange')
plt.xlabel('n')
plt.ylabel('Y(n)')
plt.title('Gráfica de Y = cos(2*pi*0.03n)')
plt.grid(True)
plt.legend()
plt.show()
```

```python

# Suma y producto entre las funciones Y1 y Y2
Ysum = Y1 + Y2
Yprod = Y1*Y2


fig, (ax1, ax2) = plt.subplots(2, 1)
# Grafica suma de las funciones
ax1.set_title('Suma')
ax1.plot(Dom, Ysum, label='Suma', color = "k")
ax1.set_xlabel('intervalo')
ax1.set_ylabel('valor')
ax1.legend(title='S(n)', loc='upper right')

# Grafica producto de las funciones
ax2.set_title('Producto')
ax2.plot(Dom, Yprod, label='Producto', color='orange')
ax2.set_xlabel('intervalo')
ax2.set_ylabel('valor')
ax2.legend(title='t(n)', loc='lower right')

# Grafica en conjunto de las 2 operaciones sobre las funciones

plt.tight_layout()
plt.show() 

plt.plot(Dom, Ysum,color = "y")
plt.plot(Dom, Yprod, color = (0.50,0.100,0.20))
plt.legend(labels = ['S(n)', 't(n)'])
plt.xlabel('n')
plt.ylabel('Valor')
plt.title('S(n) & t(n)')
plt.grid(True)
plt.show()
```

Función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.


```python
def NotasCurso(diccionario):
        
        lista_notas=[]
        
        for clave, valor in diccionario.items():            
            lista_notas.append(valor)          
            
        Nota_min = min(lista_notas)
        Nota_max = max (lista_notas)
        Media = np.mean(lista_notas)
        Desviacion = np.std(lista_notas)
        datos = [Nota_max,Nota_min,Media,Desviacion]
        serie = np.array(datos)
        return serie

print("A continuación ingrese nombre y nota de los alumnos del curso: ")
diccionario = {}
while True:
    name = input("Ingrese Nombre: ")
    grade = int(input("Ingrese nota"))
    diccionario [name] = grade
    otro = input("Pulsa 1 para finalizar, pulsa otra tecla para seguir")
    if otro == '1':
         print (diccionario)
         break

NotasCurso(diccionario)
```

```python
# Con un diccionario predefinido:

Diccionario = {"juan":23, "Pedro":32, "Ana":47, "Julieth": 38}
Serie = NotasCurso(Diccionario)
print("\nMáximo: ", Serie[0],"\nMínimo",Serie[1], "\nMedia: ",Serie[2], "\nDesviación: ",Serie[3])
```
