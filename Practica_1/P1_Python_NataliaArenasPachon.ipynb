##Item j)

```python
n = np.arange(0,101)
y = np.sin(np.pi*0.12*n)
print("y =",y)
```

##Item k)

```python
y2 = np.cos(2*np.pi*0.03*n)
print("y2 =",y2)
```

##Item l)

```python
#Suma
s = y + y2
print("Suma =",s)
```

```python
#Producto
t = y * y2
print("Producto =",t)
```

##Item m)

```python
plt.figure
plt.plot(n,y,label='y[n] = sin(π*0.12 *n)', color='#FFDAB9')
plt.plot(n, y2, label='y2[n] = cos(2π * 0.03 * n)', color='#B0E0E6')

plt.title('Señales senoidal y cosenoidal')
plt.xlabel('[n]')
plt.ylabel('Amplitud')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

##Item n)

```python
plt.figure
plt.plot(n,s,label='s[n] = y[n] + y2[n]', color='#7FFFD4')
plt.plot(n, t, label='t[n] = y[n] * y2[n]', color='#DDA0DD')

plt.title('Suma y producto de señales')
plt.xlabel('[n]')
plt.ylabel('Amplitud')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
def notas(diccionario_notas):
  notas = list(diccionario_notas.values())
  serie_notas = pd.Series(notas)
  estadisticas = pd.Series({
      'Nota mínima:':serie_notas.min(),
      'Nota máxima:':serie_notas.max(),
      'Nota media:':serie_notas.mean(),
      'Desviación típica:':serie_notas.std()
  })
  estadisticas = estadisticas.round(2)
  return estadisticas
```

```python
diccionario_notas = {'Alumno 1:':85,
                     'Alumno 2:':65,
                     'Alumno 3:':80,
                     'Alumno 4:':92,
                     'Alumno 5:':90,
                     'Alumno 6:':77,
                     'Alumno 7:':79,
                     'Alumno 8:':89,
                     'Alumno 9:':99,
                     'Alumno 10:':94,
}
resultados = notas(diccionario_notas)
print(resultados)
```
