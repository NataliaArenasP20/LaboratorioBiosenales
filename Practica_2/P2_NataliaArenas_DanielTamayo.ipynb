```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Secuencias discretas elementales**

Impulso unitario


```python
def impseq(n0,n1,n2):
  #Genera x(n) = delta(n-n0); n1<= n >= n2
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0)==0
  return[x,n]

x,n=impseq(5,-10,10)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Impulso Unitario')
plt.ylabel('$\delta[n]$')
```

**Punto 1**

Escalón unitario

```python
def escaseq(n0,n1,n2):  
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0)>=0
  return[x,n]

x,n=escaseq(0,-10,10)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Escalon Unitario')
plt.ylabel('$\mu [n]$')
```

**Punto #2**

Rampa

```python
def ramseq(n0,n1,n2):  
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = n*((n-n0)>=0)
  return[x,n]

x,n=ramseq(0,-5,15)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Rampa Discreta')
plt.ylabel('$r[n]$')
```

Exponenciales complejas

```python
def exponcencial(n0,n1,n2,img):  
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  u = (n-n0) >= 0  # Escalón unitario
  #x = np.e**(n-n0) * u
  x = np.exp(img*n) * u
  
  return[x,n]

I = 0.2 + 1j
x,n=exponcencial(5,0,25,I)
plt.stem(n,x, label='e^'+str(I)+'n')
plt.xlabel('$n$')
plt.title('Exponencial Compleja')
plt.legend()
plt.ylabel('X[n]')
```

Sinusoides


```python
def Sinusoides(n0,n1,n2,theta,omega):  
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  u = (n-n0) >= 0  # Escalón unitario
  x = np.sin(theta+omega*n)*u
  
  return[x,n]

x,n=Sinusoides(5,0,20,10,100)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Señal Sinusoidal')
plt.ylabel('X[n]')
```
