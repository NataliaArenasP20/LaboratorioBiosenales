### **Secuencias discretas elementales**

Impulso unitario


```python
def impseq(n0,n1,n2):
  #Genera x(n) = delta(n-n0); n1<= n >= n2
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0)==0
  return[x,n]

x,n=impseq(5,-10,10)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Impulso Unitario')
plt.ylabel('$\delta[n]$')
plt.grid()
```

### **Punto 1**

Escalón unitario

```python
def escaseq(n0,n1,n2):
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0)>=0
  return[x,n]

x,n=escaseq(0,-10,10)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Escalon Unitario')
plt.ylabel('$\mu [n]$')
plt.grid()
```

### **Punto #2**

Rampa

```python
def ramseq(n0,n1,n2,m):
  n=np.arange (n1,n2+1)
  r,n=escaseq(n0,n1,n2)
  x=r*(n-n0)
  return [x*m, n]

r,n = ramseq(0,-5,15,4)
plt.stem(n,r)
plt.xlabel('$n$')
plt.title('Rampa Discreta')
plt.ylabel('$r[n]$')
plt.grid()
```

### **Exponenciales complejas**

```python
def exponcencial(n0,n1,n2,img):
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  u = (n-n0) >= 0  # Escalón unitario
  #x = np.e**(n-n0) * u
  x = np.exp(img*n) * u

  return[x,n]

I = 0.2 + 1j
x,n=exponcencial(5,0,25,I)
plt.stem(n,x, label='e^'+str(I)+'n')
plt.xlabel('$n$')
plt.title('Exponencial Compleja')
plt.legend()
plt.ylabel('X[n]')
plt.grid()
```

### **Sinusoides**


```python
def Sinusoides(n0,n1,n2,theta,omega):
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  u = (n-n0) >= 0  # Escalón unitario
  x = np.sin(theta+omega*n)*u

  return[x,n]

x,n=Sinusoides(5,0,20,10,100)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Señal Sinusoidal')
plt.ylabel('X[n]')
plt.grid()
```

```python
n1 = np.arange( 0, 5+1)
n2 = np.arange( 6, 11+1)
n3 = np.arange( 12, 17+1)

[X1,n1]=ramseq(0,0,5,1)
[X2,n2]=ramseq(5,6,11,1)
[X3,n3]=ramseq(10,12,17,1)

X123 = np.concatenate((X1,X2,X3))
n123 = np.concatenate((n1,n2,n3))

plt.stem(n123,X123)
plt.xlabel('$n$')
plt.title('Secuencia')
plt.ylabel('X[n]')
plt.grid()
```

**a)** Derive la secuencia

```python
derivada_X1 = np.diff(X1)
derivada_X2 = np.diff(X2)
derivada_X3 = np.diff(X3)

derivada_X123 = np.concatenate((derivada_X1, derivada_X2, derivada_X3))
n_derivada = np.concatenate((n1[:-1], n2[:-1], n3[:-1]))

plt.stem(n_derivada, derivada_X123)
plt.xlabel('$n$')
plt.title('Derivada de la secuencia')
plt.ylabel('X[n]')
plt.grid()
```

# Descripción:
La función dada ( X[n]) es una función a tramos compuesta por 3 rampas acotadas en intervalos, por lo cuál la derivada de esta función será el diferencial individual de los tramos, lo que nos resulta en 3 escalones unitarios, 2 desplazados hacia la derecha. 

X'[n] = u[n] para  0<= n <= 5 , u[n-5] para  6<= n <= 11  , u[n-10] para  12 <= n <= 17

El resultado no corresponde con la derivada general de una función rampa debido a que x[n] tiene un Dominio finito y discontinuo, mientras que la función rampa y su derivada cuentan con dominios infinitos y continuos. Por lo que, a pesar de generarse una gráfica que podría interpretarse como una función escalón, esto sería erroneo, dado que no está definida como cero en n<0 ni como 1 en 5<n<6 , 11<n<12 , n>17 . Además, al estar graficada con el método stem, dichas discontinuidades no son evidentes a priori.
