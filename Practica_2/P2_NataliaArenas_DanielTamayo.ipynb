##Impulso unitario

```python
def impseq(n0,n1,n2):
  #Genera x(n) = delta(n-n0); n1<= n >= n2
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0)==0
  return[x,n]

x,n=impseq(0,-10,10)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Impulso Unitario')
plt.ylabel('$\delta[n]$')
```

##**Punto 1**
##Escalón unitario

```python
def escaseq(n0,n1,n2):
  #Genera x(n) = delta(n-n0); n1<= n >= n2
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = (n-n0)>=0
  return[x,n]

x,n=escaseq(0,-10,10)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Escalon Unitario')
plt.ylabel('$\mu [n]$')
```

##**Punto #2**
##Rampa

```python
def ramseq(n0,n1,n2):
  #Genera x(n) = delta(n-n0); n1<= n >= n2
  n = np.arange(n1,n2+1) #Se crea el vector de muestras
  x = n*((n-n0)>=0)
  return[x,n]

x,n=ramseq(0,-5,15)
plt.stem(n,x)
plt.xlabel('$n$')
plt.title('Rampa Discreta')
plt.ylabel('$r[n]$')
```
